```python
import numpy as np

import pandas as pd

import sympy as sp
from sympy.abc import x
from sympy import init_printing;
init_printing()

import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
%matplotlib inline
```

### Завдання 2

```python
f1 = lambda x, y: 4*x**3 - 15 + np.cos(np.sin(y))
f2 = lambda x, y: 10*x**2 + y**2 - 7 + y
x_min = -5
x_max = 5
num_0_04 = int((x_max - x_min)/0.04)+1
x_0_04 = np.linspace(x_min,x_max, num_0_04)
```

#### Табличний метод

```python
df_2 = pd.DataFrame(index=range(len(x_0_04)*len(x_0_04)))
```

```python
def fill_table_with_x(df):
    for i in range(len(x_0_04)):
        df.iloc[i*len(x_0_04):(i+1)*len(x_0_04), 0] = x_0_04[i]
        df.iloc[i*len(x_0_04):(i+1)*len(x_0_04), 1] = x_0_04[0:len(x_0_04)]
    return 0
```

```python
df_2['x1']=0
df_2['x2']=0
fill_table_with_x(df_2)
```

```python
df_2['f1(x1,x2)'] = f1(df_2['x1'], df_2['x2'])
df_2['f2(x1,x2)'] = f2(df_2['x1'], df_2['x2'])
df_2.head(10)
```

Знаходжу гарантований мінімум, який матиме 1й гравець при кожному із генерованих ним значень х1 = (-5, 4.96, -4.92, ..., 5). Тут вважаю, що 2й гравець хоче завалити 1го і надсилає йому якомога гірші х2 при будь-якому розкладі на х1 у 1го гравця (тобто якщо я 1й гравець і обираю х1=3, то який х2 дасть мені найгірший результат??? А якщо оберу х1=4.5, то тоді при якому х2 результат буде найпаскуднішим? І отак перебираю між своїми стратегіями і знаходжу для кожного значення х1 де найгірший х2 що дає найгірше значення моєї функції f1, тобто знаходжу мінімум f1 при фіксованій мною стратегії х1.). А потім обираю максимальний серед цих мінінмумів, оскільки хочу знати, який х1 обрати в своїй стратегії, аби навіть при поганих х2 від 2го гравця отримати якнайкращий результат.

```python
df_2['min_x2(f1(x1,x2))'] = 0
for i in range(len(x_0_04)):
    min_x2 = min( df_2.iloc[ np.where(df_2['x1']==x_0_04[i])[0], 2] )
    df_2.iloc[np.where(df_2['x1']==x_0_04[i])[0], 4]= min_x2
```

```python
max_x1_min_x2 = max(pd.to_numeric(df_2['min_x2(f1(x1,x2))'], errors='coerce'))
df_2['max_x1_min_x2(f1(x1,x2))']=max_x1_min_x2
```

Аналогічно.

```python
df_2['min_x1(f2(x1,x2))'] = 0
for i in range(len(x_0_04)):
    min_x1 = min( df_2.iloc[ np.where(df_2['x2']==x_0_04[i])[0], 3] )
    df_2.iloc[np.where(df_2['x2']==x_0_04[i])[0], 6]= min_x1
```

```python
max_x2_min_x1 = max(pd.to_numeric(df_2['min_x1(f2(x1,x2))'], errors='coerce'))
df_2['max_x2_min_x1(f2(x1,x2))']=max_x2_min_x1
```

#### Знаходження мінімального delta

Дельта - для кожної пари (х1, х2) знаходимо максимальний вектор нев'язки ( на скільки фактичне значення відрізняється від гарантованого(мінімального; при поганих умовах) ), і потім знаходимо таку пару (х1, х2), де ця різниця від гарантованого буде мінімальною для ОБОХ! гравців (за модулем), тобто кожен втратить мінімально і виграє максимально. Тобто дельта - оптимальний розв'язок задачі! І аби виграли обоє, дельта має прагнути до нуля.

```python
df_delta = pd.DataFrame()
```

```python
df_delta['x1'] = df_2['x1']
df_delta['x2'] = df_2['x2']
df_delta['|f1-f1*|'] = abs(df_2['f1(x1,x2)'] - f1_star)
df_delta['|f2-f2*|'] = abs(df_2['f2(x1,x2)'] - f2_star)
```

```python
df_delta
```

```python
def max_df_delta(x, y):
    returnvec=np.zeros(x.shape)
    for i in range(0, x.shape[0]):
        if x.iloc[i]<y.iloc[i]:
            returnvec[i]=y.iloc[i]
        else:
            returnvec[i]=x.iloc[i]
    return returnvec
```

Знаходимо максимум з-поміж $\Delta_i$ для кожної пари $(x_1,x_2)$.

```python
df_delta['max(delta_i)'] = max_df_delta(df_delta['|f1-f1*|'], df_delta['|f2-f2*|'])
```

```python
print df_delta.iloc[np.where(df_delta['max(delta_i)']==min(df_delta['max(delta_i)']))]
```

```python
df_delta.iloc[np.where(df_delta['max(delta_i)']==min(df_delta['max(delta_i)']))]
```
